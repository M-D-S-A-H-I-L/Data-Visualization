```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/data (1).csv')


print(df.head())
print(df.info())
print(df.describe())

print(df.isnull().sum())

df['umpire3'].fillna('Unknown', inplace=False)


```

```python
# 1. Bar plot of matches won by each team
plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=df)
plt.title('Matches Won by Each Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# 2. Distribution of player of the match awards
plt.figure(figsize=(12, 6))
sns.countplot(x='player_of_match', data=df, order=df['player_of_match'].value_counts().iloc[:10].index)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
numerical_df = df.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['season', 'winner', 'toss_winner', 'city']])
plt.show()
```

```python
#Scatterplot
sns.scatterplot(x='winner', y='season', data=df, hue='city')
plt.show()
```
