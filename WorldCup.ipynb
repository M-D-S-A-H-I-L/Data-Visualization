# Data Analysis


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/WorldCupMatches.csv')
df
# Number of rows and columns
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

```python
# Print each column name
print("Column names:")
for col in df.columns:
    print(col)
```

```python
# Print specific range using slicing (e.g., rows 10 to 20)
print("Slicing rows from index 10 to 20:")
print(df[10:21])
```

```python
# Summarization statistics
print("Summarization statistics:")
print(df.describe())
```

# Data Cleaning

```python
# Check for missing values
print("Checking for missing values:")
print(df.isnull().sum())
```

```python
# Replace missing numerical values with mean/median
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['int64', 'float64']:
            # Replace missing values with mean
            df[col].fillna(df[col].mean(), inplace=True)

df
```

# Data Visualization


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot
sns.pairplot(df.select_dtypes(include=['float64', 'int64']))
plt.show()
```

```python
#Boxplot
sns.boxplot(x='Home Team Goals', y='Away Team Goals', data=df)
plt.show()
```

```python
#Heatmap
# Check the actual column names in your DataFrame
print(df.columns)

# Drop non-numerical columns (adjust based on actual names)
numerical_df = df.select_dtypes(include=['float64', 'int64'])

corr = numerical_df.corr()
sns.heatmap(corr)
plt.show()
```

```python
# Barplot
plt.figure(figsize=(9,5))
sns.barplot(x='Home Team Name', y='Home Team Goals', data=df.head(10))
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Home Team Goals'], bins=30, kde=True)
plt.show()
```

```python
# Distribution Plot
plt.figure(figsize=(10, 6))
sns.distplot(df['Home Team Goals'], bins=30, kde=True)
plt.show()
```

```python
# Scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=df)
plt.show()
```

```python
# isnull() and isnull().sum()
# The isnull() function checks for missing values in the DataFrame.
# The isnull().sum() function provides the count of missing values for each column.
print(df.isnull().sum())

# shape
# The shape attribute returns a tuple representing the dimensionality of the DataFrame (rows, columns).
print(df.shape)

# describe()
# The describe() method provides a summary of statistics for numerical columns in the DataFrame.
print(df.describe())

# df['column_name']
# This syntax is used to select a specific column from the DataFrame.
print(df['Home Team Name'])

# max() and min()
# These functions are used to find the maximum and minimum values of a particular column.
max_value = df['Home Team Goals'].max()
min_value = df['Home Team Goals'].min()
print(f"Max value in Home Team Goals: {max_value}")
print(f"Min value in Home Team Goals: {min_value}")

```
